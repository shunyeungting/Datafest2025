```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
LeasesDB = pd.read_csv('./Data/Leases.csv')# Read Data
```

```python
print(LeasesDB.head())
```

```python
print(LeasesDB.info())
```

```python
missing_values = LeasesDB.isnull().sum()
print(missing_values)
```

```python
plt.figure(figsize=(12, 6))
sns.heatmap(LeasesDB.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()
```

```python
# Corr Matrix with following columns: RBA	available_space	availability_proportion	internal_class_rent	overall_rent
corr_matrix = LeasesDB[['RBA', 'available_space', 'availability_proportion', 'internal_class_rent', 'overall_rent']].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matri With Class O')
plt.show()
```

```python
Temp = pd.read_csv('./Data/Leases.csv')# Read Data
Temp2 = pd.read_csv('./Data/Leases.csv')# Read Data

# Binning based on RBA column
Temp['RBA'] = pd.cut(Temp['RBA'], 5)
Temp.head()

#Display bins
Temp['RBA'].value_counts()

# Binning based on leasing Column
Temp2['leasing'] = pd.cut(Temp2['leasing'], 5)
Temp2.head()

#Display bins 
Temp2['leasing'].value_counts()


# Display both on seperate histograms
plt.figure(figsize=(12, 6))
sns.histplot(Temp['RBA'], bins=5, kde=True) # Histogram for RBA
plt.title('Histogram of RBA')
plt.show()

plt.figure(figsize=(12, 6))
sns.histplot(Temp2['leasing'], bins=5, kde=True) # Histogram for Leasing
plt.title('Histogram of Leasing')
plt.show()


```
